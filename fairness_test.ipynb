```python
import pandas as pd
```

```python
columns_to_remove = [
    'Unnamed: 0',
    'id_student', 
    'id_student_original',
    'id_year',
    'id_grade',
    'id_class_group',
    'id_school',
    'id_student_16_19',
    'id_school_16_19',
    'student_questionnaire',
    'principals_questionnaire',
    'principals_questionnaire', 
    'teachers_questionnaire', 
    'census',
    'family_questionnaire'
]
```

```python
dataset = pd.read_csv('./Dataset.csv')
```

```python
dataset = dataset.drop(columns=columns_to_remove)
```

```python
data_dictionary = pd.read_csv('./Data_dictionary.csv')
```

```python
data_dictionary
```

```python
categorical_attributes = data_dictionary[data_dictionary['Unnamed: 2'] == 'Categorical']
```

```python
continuous_attributes = data_dictionary[data_dictionary['Unnamed: 2'] == 'Continuous']
```

```python
categorical_attributes['Unnamed: 0']
```

```python
for column in dataset.columns:
    if column in categorical_attributes['Unnamed: 0'].values:
        dataset[column].fillna(float(dataset[column].mode()), inplace=True)
    elif column in continuous_attributes['Unnamed: 0'].values:
        print("strunz")
        dataset[column].fillna(float(dataset[column].mean()), inplace=True)
```

```python
continuous_attributes['Unnamed: 0']
```

```python
dataset.to_csv('./ULL_dataset.csv')
```
