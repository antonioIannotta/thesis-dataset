```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
columns_to_remove = [
    'Unnamed: 0',
    'id_student', 
    'id_student_original',
    'id_year',
    'id_grade',
    'id_class_group',
    'id_school',
    'id_student_16_19',
    'id_school_16_19',
    'student_questionnaire',
    'principals_questionnaire',
    'principals_questionnaire', 
    'teachers_questionnaire', 
    'census',
    'family_questionnaire'
]
```

```python
dataset = pd.read_csv('./Dataset.csv')
```

```python
dataset_dictionary = pd.read_csv('./Data_dictionary.csv')
```

```python
dataset = dataset.drop(columns=columns_to_remove)
```

```python
for column in dataset.columns:
    dataset.rename(columns={column: column.lower()}, inplace=True)
```

```python
categorical_attributes_dataframe = dataset_dictionary[dataset_dictionary['Unnamed: 2'] == 'Categorical']
```

```python
categorical_attributes = categorical_attributes_dataframe['Unnamed: 0']
```

```python
categorical_attributes
```

```python
for categorical_attribute in categorical_attributes:
    #fill with mode of the categorical attribute
    mode = dataset[categorical_attribute.lower()].mode()
    dataset[categorical_attribute.lower()].fillna(mode.values[0], inplace=True)
```

```python
continuous_attributes_dataframe = dataset_dictionary[dataset_dictionary['Unnamed: 2'] == 'Continuous']
```

```python
continuous_attributes = continuous_attributes_dataframe['Unnamed: 0']
```

```python
continuous_attributes
```

```python
for continuous_attribute in continuous_attributes:
    mean = dataset[continuous_attribute.lower()].mean()
    dataset[continuous_attribute.lower()].fillna(mean, inplace=True)
```

```python
dataset
```

```python
def categorical_to_numeric_converter(dataset: pd.DataFrame) -> pd.DataFrame:
    """
    The method converts each categorical column into a numerical one
    :param dataset: the dataset on which pervorm the convertion
    :return: returns a dataset with all numerical attributes
    """
    categorical_columns = [column for column in dataset.columns if dataset[column].dtype=="O"]
    label_encoder = LabelEncoder()
    for column in categorical_columns:
        dataset[column] = label_encoder.fit_transform(dataset[column])

    return dataset
```

```python
dataset = categorical_to_numeric_converter(dataset)
```
